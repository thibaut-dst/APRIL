```python
from collections import Counter
import re
```

## Basic method - small files
### .txt

```python
# Read text file
with open('text-detection.txt', 'r') as file:
    text = file.read()

# Tokenize and normalize text
words = re.findall(r'\b\w+\b', text.lower())
display(words)
```

```python
# Count word frequencies
word_count = Counter(words)

# Detect specific words
target_words = ['to', 'machine']
for word in target_words:
    print(f"{word}: {word_count[word]}")
```

### .html

```python
# Read text file
with open('text-in-html.html', 'r') as file2:
    text2 = file2.read()

# Tokenize and normalize text
words2 = re.findall(r'\b\w+\b', text2.lower())
display(words2)
```

```python
# Count word frequencies
word_count2 = Counter(words2)

# Detect specific words
target_words2 = ['div', 'fichier', 'Word sur votre ordinateur']
for word2 in target_words2:
    print(f"{word2}: {word_count2[word2]}")
```

### Advanced method - Large files 10+ Mb & NLP features

```python
import spacy
nlp = spacy.load("fr_dep_news_trf")
```

```python
with open('text-detection.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# Process the text with spaCy
doc = nlp(text)
display(doc)
```

```python
# Define target words to detect (you can adjust this list)
target_words = ['format', 'fichiers', 'des']

# Create a dictionary to count occurrences
word_count = {word: 0 for word in target_words}

# Tokenize the text and check for specific words
for token in doc:
    if not token.is_punct:  # Exclude punctuation
        if token.text.lower() in word_count:  # Check if the word is in our list
            word_count[token.text.lower()] += 1

# Print out the word frequencies
for word, count in word_count.items():
    print(f"'{word}' appears {count} times in the text.")
```

## Additional Text Processing

- Lemmatize the words (get their base form), so "scientifique" and "scientifiques" are treated as the same word.
- Part-of-speech tagging to detect whether a word is used as a noun, verb, etc.
- Named Entity Recognition (NER) to find named entities like "Paris" or "France."


### Lemmatize

```python
for token in doc:
    if not token.is_punct:  # Exclude punctuation
        print(f"{token.text} -> {token.lemma_}")  # Print word and its lemma
```

### Named Entity Recognition

```python
# Perform Named Entity Recognition (NER)
for ent in doc.ents:
    print(f"{ent.text} - {ent.label_}")
```
